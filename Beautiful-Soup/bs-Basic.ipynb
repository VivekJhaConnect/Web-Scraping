```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + "iPhone12Pro"
flipkart_url
```

```python
import ssl

# Disable SSL certificate verification
ssl_context = ssl.create_default_context()
ssl_context.check_hostname = False
ssl_context.verify_mode = ssl.CERT_NONE
```

```python
urlclient = urlopen(flipkart_url, context=ssl_context)
```

```python
flipkart_page = urlclient.read()
flipkart_page
```

```python
flipkart_html = bs(flipkart_page, 'html.parser')
flipkart_html
```

```python
bigbox = flipkart_html.find_all('div',{"class":"cPHDOP col-12-12"})
```

```python
del bigbox[0:3]
```

```python
productlink = "https://www.flipkart.com" + bigbox[0].div.div.div.a['href']
```

```python
product_request = requests.get(productlink)
```

```python
product_html = bs(product_request.text, 'html.parser')
```

```python
comment_box = product_html.find_all('div', {'class':'RcXBOT'})
comment_box
```
